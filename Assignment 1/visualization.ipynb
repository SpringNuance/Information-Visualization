```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python


percent_values = [22, 26, 39, 10, 3]
colors = ['blue', 'yellow', 'green', 'pink', 'red']
labels = ["under\n 100\$", "about\n 100\$", "100\$ to\n 200\$", '200\$ to \n300\$', 'more than \n \$300']

#print(sns.color_palette('Set1'))
colors = [(0.21568627450980393, 0.49411764705882355, 0.7215686274509804), (1.0, 0.4980392156862745, 0.0),(0.30196078431372547, 0.6862745098039216, 0.2901960784313726), (0.596078431372549, 0.3058823529411765, 0.6392156862745098), (0.8941176470588236, 0.10196078431372549, 0.10980392156862745)]

#sns.set(font_scale = 1.3)
plt.figure(figsize=(8, 8))
plt.title("How much do you spend on groceries every week?", fontsize=20)

patches, texts, autotexts = plt.pie(
    x=percent_values, 
    labels=labels,
    autopct='%1.0f%%',
    startangle=90,
    colors=colors,
    counterclock=False, 
    labeldistance=1.15,
    # Distance of percent labels from the center
    pctdistance=0.8,
)

# Customize text labels
for text in texts:
    text.set_horizontalalignment('center')

# Customize percent labels
for autotext in autotexts:
    autotext.set_horizontalalignment('center')
    autotext.set_fontstyle('italic')

plt.savefig('figure1.png')
plt.show()
```

```python
import pandas as pd
#import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime

plt.figure(figsize=(12, 5), frameon=False)
plt.style.use('dark_background')
#plt.tight_layout()
plt.grid(False)
# read the CSV file into a pandas DataFrame
df = pd.read_csv('BTCvsSP500.csv')

# rename the column
df = df.rename(columns={'Unnamed: 0': 'index'})
#print(df.columns)
# plot the time series using seaborn
index = df['index']
date = df['date']
BTC = df['BTC']
SP500 = df['SP500']

# define a lambda function to convert the date string
#date_format = lambda x: datetime.strptime(x, '%Y-%m-%d').strftime('%d %b').replace(' 0', ' ')
date_format = lambda x: datetime.strptime(x, '%Y-%m-%d').strftime('%dth\n%b').replace(' 0', ' ')
# apply the lambda function to the pandas series
formatted_dates = date.apply(date_format)
#print(formatted_dates)
plt.title("BITCOIN IS THE FUTURE PREMISE\nA promising stable replacement for traditional stocks", fontsize=20)

plt.ylabel("Value ($)", fontsize=16)
#plt.xlabel("Day", fontsize=16)
# change the font size of the x-axis ticks
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.ylim([2000, 20000])
plt.plot(formatted_dates[1200:1218], BTC[1200:1218], color='blue', label="BTC")
plt.plot(formatted_dates[1200:1218], SP500[1200:1218], color='orange', label="S&P500")
plt.legend(fontsize=16)
#plt.show()

plt.savefig('figure2.png', transparent=True)
```
